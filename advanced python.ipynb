```python
##try else


try:
    
    ##code block where exception can occur
    a=int(input('enter number 1'))
    b=int(input('enter number 2'))
    c=a/b
    d=a*b
    e=a+b
   


except NameError as ex1:
    print('the user have not defined the variable')

except TypeError:
    print('try to make datatype similar ')
    
except Exception as ex:
    print(ex)

else:
    print(c)
    print(d)
    print(e)
```

```python
## try else and finally


try:
    
    ##code block where exception can occur
    a=int(input('enter number 1'))
    b=int(input('enter number 2'))
    c=a/b
    d=a*b
    e=a+b
   


except NameError as ex1:
    print('the user have not defined the variable')

except TypeError:
    print('try to make datatype similar ')
    
except Exception as ex:
    print(ex)

else:
    print(c)

finally:
    print('the execution is done')
```

```python
## closures

def main_welcome(msg):
    def sub_welcome_class():
        print('Welcome to the lions den')
        print(msg)  # Accessing the enclosed 'msg'
        print('Jackal is hunted')
    return sub_welcome_class  # Return the function itself, not its result

welcome_function = main_welcome('A custom message') 
welcome_function()  # Now call the returned function
```

```python
main_welcome(print)
```

```python
## initial decorators

def main_welcome(func):
    def sub_welcome_class():
        print('Welcome to the lions den')
        print(func ([1,2,3,45] ))  # Call the decorated function without arguments
        print('Jackal is hunted')
    return sub_welcome_class 

main_welcome(len)
```

```python
def main_welcome(func):
    def sub_welcome_class(*args, **kwargs):  # Accept any arguments
        print('Welcome to the lions den')
        result = func(*args, **kwargs)  # Call func with provided arguments
        print(result)
        print('Jackal is hunted')
    return sub_welcome_class 

@main_welcome  # Apply the decorator to a function
def my_function(numbers):
    return len(numbers)

my_function([1, 2, 3, 45])
```

```python
from numba import jit 
import numpy as np 
import time

x=np.arrange(100).reshape(10,10)

@jit(nopython=True)
def go_fast(a):
    trace=0.0
    for i in range(a.shape[0]):
        trace+= np.tanh(a[i,i])
    return a + trace 

start=time.time()
go_fast(x)
end=time.time()
print('elapsed(with compilation) = %s' % (end- start) )

start=time.time
go_fast(x)
end=time.time
print('elapsed(with compilation) = %s' % (end- start) )

```
